```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("contours.png")
img_trans = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(2,2,1)
plt.title("Transformed"), plt.imshow(img_trans), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2)
plt.title("Gray"), plt.imshow(img_gray,cmap='gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
edges = cv2.Canny(img_gray,30,200)
plt.imshow(edges,cmap='gray'),plt.xticks([]),plt.yticks([])
```

```python
contours, hierarchy = cv2.findContours(edges,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
plt.imshow(edges,cmap='gray'),plt.xticks([]),plt.yticks([])
```

```python
print("Number of Contours found = " + str(len(contours)))
```

```python
cv2.drawContours(img,contours,-1,(0,255,0),5)
plt.imshow(img),plt.xticks([]),plt.yticks([])
```
