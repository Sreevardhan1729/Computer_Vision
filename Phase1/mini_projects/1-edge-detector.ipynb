```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("edge.png")
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(img), plt.xticks([]),plt.yticks([])
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
edges = cv2.Canny(img,175,250,5,L2gradient=True)
```

```python
plt.figure()
plt.title("edges")
plt.imsave("edge-detected.png",edges,cmap='gray',format='png')
plt.imshow(edges,cmap='gray')
plt.show()
```

```python
smooth = cv2.bilateralFilter(img,d=9,sigmaColor=75,sigmaSpace=75)
edges_inv = cv2.bitwise_not(edges)
edges_inv_colored = cv2.cvtColor(edges_inv,cv2.COLOR_GRAY2BGR)
cartoon = cv2.bitwise_and(smooth,edges_inv_colored)
plt.subplot(2,2,1)
plt.title("Smooth"), plt.imshow(smooth,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(2,2,2)
plt.title("edges_inv"),plt.imshow(edges_inv,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(2,2,3)
plt.title("edges_inv_colored"),plt.imshow(edges_inv_colored),plt.xticks([]),plt.yticks([])
plt.subplot(2,2,4)
plt.title("cartoon"),plt.imshow(cartoon,cmap='gray'),plt.xticks([]),plt.yticks([])
```
